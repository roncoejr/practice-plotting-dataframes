```python
import mysql.connector
import os
import pandas as pd
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
learn_db = mysql.connector.connect(host = os.getenv("hostname"), user = os.getenv("username"), passwd = os.getenv("password"), database = "coejr_learning")
```

```python
print(learn_db)
```

```python
the_sql_gas = "select * from tbl_gasbill order by transactionDate"
```

```python
the_sql_elec = "select * from tbl_electricbill order by transactionDate"
```

# Create a dataframe from the MySQL database

```python
df_gas = pd.read_sql(the_sql_gas, learn_db, parse_dates=['transactionDate'])
```

```python
df_elec = pd.read_sql(the_sql_elec, learn_db, parse_dates=['transactionDate', 'dueDate'])
```
